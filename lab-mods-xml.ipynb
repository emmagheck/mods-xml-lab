# Lab: working with MODS XML & XPath

Part 1: count the individual MODS records in the file

```python
import xml.etree.ElementTree as ET
```

```python
mods = ET.parse(open('/Users/emmaheck/Documents/umich/courses/SI676/networked-services-labs-2023-main/data/xml/2018_lcwa_MODS_25.xml'))

root = mods.getroot()
```

```python
root.tag
```

```python
type(mods)
```

```python
for item in root:
    print(item.tag)
```

```python
namespace = {
    'mods': 'http://www.loc.gov/mods/v3'
}
```

```python
namespace['mods']
```

```python
for identifier in root.findall('.//mods:identifier[@type]', namespaces=namespace): 
    print(identifier.tag, identifier.attrib)
```

```python
# number of MODS records
mods_records =  root.findall('.//mods:mods', namespaces=namespace)
```

```python
num_records = len(mods_records)
print(num_records)
```

Part 2: Look through subject tags, identify only the ones that include an LCSH attribute, then print the content of those subject headings. Use an XPath expression here

```python
for item in root.iter():
    print(item.tag)
```

```python
subject_tags = root.findall(".//mods:subject[@authority='LCSH']", namespace)
for tag in subject_tags:
    print(tag.tag, tag.attrib)
```

Part 3: check the local call number references to ensure that they are in the proper format (e.g., lcwaAddddddd) using a regular expression.

```python
import re
```

```python
call_num_regex = re.compile(r'[a-z]{4}[A-Z]\d{7}')
```

```python
call_num_list = re.findall(call_num_regex, open('/Users/emmaheck/Documents/umich/courses/SI676/networked-services-labs-2023-main/data/xml/2018_lcwa_MODS_25.xml').read())

for call_num in call_num_list:
    print(call_num)
```

Part 4: identify the local call numbers, then check to make sure all of them have appropriate attribute data attached. Make sure that there is a type attribute as well as an updated attribute

```python
for call_num in call_num_list:
    call_num.set('local', 'yes')
    print(call_num.tag, call_num.attrib)
```
